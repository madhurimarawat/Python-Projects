```python
# Importing required Libraries

# Importing pandas to read csv file and for data manipulation
import pandas as pd

# Importing matplotlib to provide axis labels and titles to the plot
import matplotlib.pyplot as plt

# Importing seaborn for plotting
import seaborn as sns

# Importing csv file using pandas
df=pd.read_csv("/content/Automobile.csv")
df
```

```python
# Seeing first 7 rows of dataframe using head() function
df.head(7)
```

```python
# Plotting line graph between Length and Mileage with engine as the hue parameter(to understand the variation)
# We can see that cars with lesser length are giving more mileage and cars with more length are giving lesser mileage and the highest mileage is given by ohc engine
sns.lineplot(df,x=df["length"],y=df["mileage"],hue=df["engine"])
plt.title("Length and Mileage of Automobiles wrt engine")
plt.xlabel("Length of the Automobile")
plt.ylabel("Mileage given by the Automobile")
```

```python
# Plotting a scatter plot to visualize the data in a more clear way
# All points are visible in this plot
# Plotting scatter plot between Length and Mileage with engine as the hue parameter(to understand the variation)
# We can see that cars with lesser length are giving more mileage and cars with more length are giving lesser mileage and the highest mileage is given by ohc engine
sns.scatterplot(df,x=df["length"],y=df["mileage"],hue=df["engine"])
plt.title("Length and Mileage of Automobiles wrt engine")
plt.xlabel("Length of the Automobile")
plt.ylabel("Mileage given by the Automobile")
```

```python
# Now let's make a regression or correlation plot to understand the correlation between length and mileage
# With the plot it is clear that their is a negative correlation between length and mileage which implies smaller(less) length cars have more mileage and vice a versa
sns.regplot(x='length',y='mileage',data=df)
plt.title("Length and Mileage of Automobiles wrt engine")
plt.xlabel("Length of the Automobile")
plt.ylabel("Mileage given by the Automobile")
```

```python
# Plotting scatter plot between Length and Mileage with sunroof as the hue parameter(to understand the variation)
# We can see that most longer cars have sunroof while smaller or compact cars do not have sunroof
# Also few cars are there which gives good mileage , are compact and also have sunroof
# Thus we can  see their is no strong relation between cars having sunroof on mileage
sns.scatterplot(df,x=df["length"],y=df["mileage"],hue=df["sunroof"])
plt.title("Length and Mileage of Automobiles wrt sunroof")
plt.xlabel("Length of the Automobile")
plt.ylabel("Mileage given by the Automobile")
```

```python
# Plotting scatter plot between Length and Mileage with body-style as the hue parameter(to understand the variation)
# We can see that hatchback usually come in compact(smaller) size giving good mileage while sedan and wagon provide the most variation as they come in both smaller and larger length cars and also gives good mileage.
# Convertible also gives good mileage in medium length
# hardtop usually comes in longer length and lesser mileage
sns.scatterplot(df,x=df["length"],y=df["mileage"],hue=df["body-style"])
plt.title("Length and Mileage of Automobiles wrt body-style")
plt.xlabel("Length of the Automobile")
plt.ylabel("Mileage given by the Automobile")
```

```python
# Plotting relation plot between Length and Mileage with body-style and sunroof as the hue parameter(to understand the variation)
# From the plot we can see that sedans body style contains the most sunroof
# That is we can say that most sedans come with sunroof
# Thus we can see that though body style has effect on mileage ,their is no such case with sunroof
sns.relplot(x='length',y='mileage',hue=df[['body-style','sunroof']].apply(tuple,axis=1),data=df)
plt.title("Length and Mileage of Automobiles wrt body-style and sunroof")
plt.xlabel("Length of the Automobile")
plt.ylabel("Mileage given by the Automobile")
```

```python
# Plotting relation plot between Length and Mileage with body-style and sunroof as the hue parameter(to understand the variation)
# From the plot we can see that sedans is the most popular body style and is manufactured by most companies from luxury brands to normal brands
# hatchback and wagon body style are also manufactured by most companies
# convertible and hardtop is manufactured by few companies as they are mostly luxury vehicles made by high end companies like porsche or mercedes
sns.relplot(x='length',y='mileage',hue=df[['body-style','company']].apply(tuple,axis=1),data=df)
plt.title("Automobiles wrt body-style and company")
plt.xlabel("Length of the Automobile")
plt.ylabel("Mileage given by the Automobile")
```

```python
# Plotting relation plot between Length and Mileage with body-style, company and sunroof as the hue parameter(to understand the variation)
# From the plot we can see that sedans is the most popular body style and is manufactured by most companies from luxury brands to normal brands and also most sedans contain sunroof
# hatchback and wagon body style are also manufactured by most companies and they usually dosen't have sunroof
# convertible and hardtop is manufactured by few companies as they are mostly luxury vehicles made by high end companies like porsche or mercedes and they always have sunroof
sns.relplot(x='length',y='mileage',hue=df[['body-style','company','sunroof']].apply(tuple,axis=1),data=df)
plt.title("Automobiles wrt body-style,company and sunroof")
plt.xlabel("Length of the Automobile")
plt.ylabel("Mileage given by the Automobile")
```
