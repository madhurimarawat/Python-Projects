### This First Code is for Simple one player, playing against computer for an arbitary n number of rounds.

### Conditions:

- Number of rounds n will be given by user and it should be strictly odd(To Decide Winner) else we will convert it to odd.

- User can choose their name.

- Final Result will be the sum of all rounds.

```python
import random

# ANSI escape codes for text formatting and color
GREEN = '\033[92m'
RED = '\033[91m'
RESET = '\033[0m'
BLOCK = '\033[34m'

# Text formatting
BOLD = '\033[1m'
ITALIC = '\033[3m'
UNDERLINE = '\033[4m'

# Colors
BLACK = '\033[30m'
RED = '\033[31m'
GREEN = '\033[32m'
YELLOW = '\033[33m'
BLUE = '\033[34m'
MAGENTA = '\033[35m'
CYAN = '\033[36m'
WHITE = '\033[37m'

# Background colors
BG_BLACK = '\033[40m'
BG_RED = '\033[41m'
BG_GREEN = '\033[42m'
BG_YELLOW = '\033[43m'
BG_BLUE = '\033[44m'
BG_MAGENTA = '\033[45m'
BG_CYAN = '\033[46m'
BG_WHITE = '\033[47m'

# Function for Rock, Paper, and Scissors
def computeScore(choice1 , choice2):

    # For Draw condition no one will be having any score
    if choice1 == choice2:
      print("It's a draw! Both sides have same choice. ğŸ‘”")
      return 0

    # For Different Choices
    elif (choice1 == "P" and choice2 == "R") or (choice1 == "R" and choice2 == "S") or (choice1 == "S" and choice2 == "P"):

      print(GREEN)
      if choice1 == "P" and choice2 == "R":
        print("User Paper ğŸ“„ Covered Computer Rock ğŸª¨ Completely!!!")
        print("Congratulations ğŸ§£ğŸ‰ User")

      elif choice1 == "R" and choice2 == "S":
        print("User Rock ğŸª¨ Crushed Computer Scissors âœ‚ï¸ Completely!!!")
        print("Congratulations ğŸ’ªğŸ‰ User")

      else:
        print("User Scissors âœ‚ï¸ Cut Computer Paper ğŸ“„ Completely!!!")
        print("Congratulations âš”ï¸ğŸ‰ User")

      print(RESET)

      return 1

    # Else Second Player wins
    else:

      print(RED)

      if choice1 == "P" and choice2 == "R":
        print("Computer Paper ğŸ“„ Covered User Rock ğŸª¨ Completely!!!")
        print("Better luck next time! ğŸ¤–ğŸ’”")

      elif choice1 == "R" and choice2 == "S":
        print("Computer Rock ğŸª¨ Crushed User Scissors âœ‚ï¸ Completely!!!")
        print("Better luck next time! ğŸ¤–ğŸ’”")

      else:
        print("Computer Scissors âœ‚ï¸ Cut User Paper ğŸ“„ Completely!!!")
        print("Better luck next time! ğŸ¤–ğŸ’”")

      print(RESET)

    return 2


# Now we will have a function to add score
def calculateScore(choice1 , choice2, result):

  # Check if the user input is valid (S, R, or P)
  if choice2.upper() not in ["S", "R", "P"]:

    print(f"{RED}Invalid input! Computer scores this round. ğŸ’»ğŸ’”{RESET}")

    result.append([0, 1])
    return result

  # List to score result
  score = computeScore(choice1 , choice2)

  if score == 0:
    result.append([0,0])

  elif score == 1:
    result.append([1,0])

  else:
    result.append([0,1])

  # Returning Result
  return result


# Function to calculate and print final scores
def finalScore(result):
    user_score = sum([round_result[0] for round_result in result])
    computer_score = sum([round_result[1] for round_result in result])

    print("Game Over! Final Scores:")
    print(f"{YELLOW}User: {user_score} | Computer: {computer_score}{RESET}\n")

    if user_score > computer_score:
        print(f"{GREEN}Congratulations! You are the overall winner! ğŸ‰ğŸŒŸğŸ‰{RESET}")
    elif user_score < computer_score:
        print(f"{RED}Better luck next time! Computer is the overall winner! ğŸ¤–ğŸ’”{RESET}")
    else:
        print("It's a draw! Both sides are equally matched. ğŸ‘”")

    print()
    play_again = input("Enter 1 if you want to continue with another game, else press any key: ")

    if play_again == "1":

        print()
        print("Let the games continue! ğŸ®ğŸ”¥\n\n")
        main()
    else:
        print("\nThanks for playing! Have a great day! ğŸ˜ŠğŸ‘‹")

# Main Function
def main():
    choices = ["R", "P", "S"]
    res = []

    print(f"{BLOCK}{BOLD}{'ğŸª¨ğŸ“„âœ‚ï¸' * 10} {RESET}{RESET}\n")
    print(f"{BLOCK}{BOLD}{ITALIC}The Quirky Quest of ğŸª¨ Rock, ğŸ“„ Paper, and âœ‚ï¸ Scissors Delight{RESET}{RESET}{RESET}\n")
    print(f"{BLOCK}{BOLD}{'ğŸª¨ğŸ“„âœ‚ï¸' * 10} {RESET}{RESET}\n")
    print(f"{YELLOW}{BOLD}{ITALIC} Enter Your Choice and Enjoy ğŸ‰!!!{RESET}{RESET}{RESET}\n")

    n = int(input("Enter the Number of Rounds: "))
    print()

    for i in range(n):
        choice = random.choice(choices)
        print(f"{MAGENTA}{BOLD}{UNDERLINE} {'*' * (i + 1)} Round {i + 1} {'*' * (i + 1)}{RESET}{RESET}{RESET}")
        print()

        try:
            print(f"{CYAN}{BOLD}{ITALIC} Enter R for Rock ğŸª¨, P for Paper ğŸ“„, and S for Scissors âœ‚ï¸: {RESET}{RESET}{RESET}\n")
            userChoice = input()
            print()

        except:
            print(f"{RED}{BOLD}{ITALIC}{UNDERLINE}Please Enter Correct Input!! ğŸš«{RESET}{RESET}{RESET}{RESET}\n")
            continue

        res = calculateScore(choice, userChoice, res)
        print()

    finalScore(res)

if __name__ == "__main__":
    main()
```
